### 6.2.4 	示例：写入水果数据到CSV文件

```python
data1 = { 'apples': [3, 2, 0, 1],
          'oranges': [0, 1, 2, 3],
          'bananas': [1, 2, 1, 0] }

df1 = pd.DataFrame(data1, index=['June','Robert','Lily','David'])
df1
```

```python
df1.to_csv(file_path+'水果.csv', index=False)
```

```python
df1.to_csv(file_path+'水果.csv', header=False)
```

```python
df1.to_csv(file_path+'水果.csv', header=['苹果','桔子','香蕉'])
```

```python
df1.to_csv(file_path+'水果.csv', header=['苹果','桔子','香蕉'],encoding='gbk')
```
