### 6.1.4 	示例：写入水果数据到Excel文件

```python
data1 = { 'apples': [3, 2, 0, 1],
          'oranges': [0, 1, 2, 3],
          'bananas': [1, 2, 1, 0] }

df1 = pd.DataFrame(data1, index=['June','Robert','Lily','David'])
df1
```

```python
df1.to_excel(file_path+'水果.xlsx')
```

```python
df1.to_excel(file_path+'水果.xlsx', index=False)
```

```python
df1.to_excel(file_path+'水果.xlsx', sheet_name='第1天')
```

```python
df1.to_excel(file_path+'水果.xlsx', sheet_name='第1天', header=False)
```

```python
df1.to_excel(file_path+'水果.xlsx', sheet_name='第1天', header=['苹果','桔子','香蕉'])
```

```python
apples = pd.Series([3,2,0,1], index=['a','b','c','d'])
apples
```

```python
apples.to_excel(file_path+'apples.xlsx')
```

```python
with pd.ExcelWriter(file_path+'水果.xlsx') as writer:
    df1.to_excel(writer, sheet_name='第1天')
    apples.to_excel(writer, sheet_name='第2天')
```
